```python
import json
import os
import datetime
import glob
```

```python
def alljson(sensor):
    alldata = {}
    start_time = datetime.datetime(2021, 12, 17, 0)
    stop_time = datetime.datetime(2022, 1, 15, 23)
    hours = (stop_time - start_time)/datetime.timedelta(hours = 1)
    hours = int(hours)
    for hour in range(hours):
        start_time =  start_time.strftime('%Y-%m-%d-%H')
        path = './sensor_data/' + sensor + '/' + start_time + '.json'
        try:
            if os.path.exists(path):
                print(path)
                json_file = open(path)
                data_loaded = json.load(json_file)
                alldata.update(data_loaded)
        except:
            pass
            
        start_time = datetime.datetime.strptime(start_time, '%Y-%m-%d-%H')
        start_time = start_time + datetime.timedelta(hours = 1)
    
    with open("./sensor_data/" + sensor + '/all.json', 'w') as f:
        json.dump(alldata, f, indent=4)
```

```python
def allsensor():
    for sensor_number in range(1, 22):
        sensor = "sensor" + str(sensor_number)
        alljson(sensor)
```

```python
allsensor()
```
